## 2. Type Annotations and Inference

### Basic Concepts
TypeScript offers two ways to work with types:
- **Explicit Typing**: You explicitly declare the type of a variable
- **Type Inference**: TypeScript automatically determines the type based on the assigned value

When to use each:
- Use explicit types for: function parameters, return types, object literals, when initial value might change
- Use type inference for: simple variables with immediate assignment, when type is obvious

```typescript
// Explicit Type Annotations
let greeting: string = "Hello, TypeScript!";
let userCount: number = 42;
let isLoading: boolean = true;
let scores: number[] = [100, 95, 98];

// Output the values
console.log(greeting);
console.log(userCount);
console.log(isLoading);
console.log(scores);

// Function with explicit parameter and return types
function greetUser(name: string): string {
    return `Hello, ${name}!`;
}

// TypeScript will ensure you pass the correct argument type
greetUser("Alice"); // OK
// greetUser(42);     // Error: Argument of type '42' is not assignable to parameter of type 'string'
```

### Type Inference Examples

TypeScript can automatically infer types from values:

```typescript
// TypeScript infers 'string'
let username = "alice";

// TypeScript infers 'number'
let score = 100;

// TypeScript infers 'boolean[]'
let flags = [true, false, true];

// TypeScript infers return type as 'number'
function add(a: number, b: number) {
    return a + b;
}
```

## 4. Arrays

### Basic Array Operations

```typescript
// 1. Initialization and push
const names: string[] = [];
names.push("Dylan");    // no error
console.log(names);
// names.push(56);      // Error: number can't be assigned to parameter of type string

// 2. Readonly arrays - changes cannot be made
const names2: readonly string[] = [];
// names2.push("Dylan");    // Error: readonly array

// 3. Type inference
const numbers = [1, 2, 3]; // inferred to type number[]
numbers.push(4); // no error
// numbers.push("2"); // Error: Argument of type 'string' is not assignable to parameter of type 'number'

let head: number = numbers[0]; // no error
console.log(head);
```

## 5. Functions

### Return Types
Functions should have explicit return types for type safety:

```typescript
// The `: number` here specifies that this function returns a number
function getTime(): number {
  return new Date().getTime();
}

console.log(getTime());
```

### Function Parameters
Type-safe function parameters ensure correct argument passing:

```typescript
function printCoord(pt: { x: number; y: number }): void {
  console.log("The coordinate's x value is " + pt.x);
  console.log("The coordinate's y value is " + pt.y);
}
```

### Optional Parameters and Default Values

```typescript
// Optional parameter
function greetOptional(name?: string): void {
  if (name) {
    console.log(`Hello, ${name}!`);
  } else {
    console.log("Hello!");
  }
}

// Default parameter
function greetWithDefault(name: string = "Guest"): void {
  console.log(`Hello, ${name}!`);
}
```

### Rest Parameters
Rest parameters allow functions to accept variable number of arguments:

```typescript
function sum(...numbers: number[]): number {
  return numbers.reduce((a, b) => a + b, 0);
}

console.log(sum(1, 2, 3, 4, 5)); // 15
```

## 7. Type Aliases

Type aliases allow you to create custom, reusable names for any data type:

```typescript
type CarYear = number;
type CarType = string;
type CarModel = string;
type Car = {
  year: CarYear;
  type: CarType;
  model: CarModel;
};

const carYear: CarYear = 2001;
const carType: CarType = "Toyota";
const carModel: CarModel = "Corolla";
const car: Car = {
  year: carYear,
  type: carType,
  model: carModel
};

console.log(car);
```

## 10. Inheritance

Classes can extend other classes and implement interfaces:

### Implement vs Extends
- **implement**: follow the same structure
- **extends**: inherit from parent class

```typescript
interface Shape {
  getArea: () => number;
}

// Implement interface
class Rectangle implements Shape {
  public constructor(protected readonly width: number, protected readonly height: number) {}

  public getArea(): number {
    return this.width * this.height;
  }
}

// Extend class
class Square extends Rectangle {
  public constructor(width: number) {
    super(width, width);
  }

  // Override method
  public override toString(): string {
    return `Square[width=${this.width}]`;
  }
}
```

## 13. Special Types

### Unknown
The type-safe counterpart of `any`. Must be type-checked before use:

```typescript
// Key differences between unknown and any:
// - unknown must be type-checked before use
// - You can't access properties on an unknown type without type assertion
// - You can't call or construct values of type unknown

let unknownValue: unknown;
unknownValue = "hello";

// Type assertion needed
if (typeof unknownValue === 'string') {
    console.log(unknownValue.toUpperCase());
}
```

## 14. Generics

Generics allow you to create reusable components with variable types:

```typescript
function createPair<S, T>(v1: S, v2: T): [S, T] {
    return [v1, v2];
}

console.log(createPair<string, number>('hello', 42)); // ['hello', 42]
```

### Generic Classes

```typescript
class Container<T> {
    private value: T;

    constructor(value: T) {
        this.value = value;
    }

    getValue(): T {
        return this.value;
    }
}

const stringContainer = new Container<string>("Hello");
const numberContainer = new Container<number>(42);
```

### Generic Constraints

Constrain generic types to extend specific types:

```typescript
function getProperty<T, K extends keyof T>(obj: T, key: K) {
    return obj[key];
}

const person = { name: "John", age: 30 };
const name = getProperty(person, "name"); // OK
// const invalid = getProperty(person, "email"); // Error: "email" is not a key of person
```

### Generic Default Values

```typescript
type Maybe<T = string> = T | null;

const maybeString: Maybe = null; // null | string
const maybeNumber: Maybe<number> = null; // null | number
```


# 15 Utility Types


## PArtial -> changes all the properties in an object to be optional

```typescript
interface Point {
    x: number;
    y: number;
}

let pointPart: Partial<Point> = { x: 10 }; // y is optional
//or pointPart.x = 10; // y is optional
```

## REQUIRED-> all prop are required

```typescript
interface Car {
    make: string;
    model: string;
    mileage ?: number;
}
let myCar: Required<Car> = { make: "Toyota", model: "Corolla", mileage: 50000 }; // mileage is required now
```

## RECORD -> SHORTCUT TO DEFINE AN OBJECT TYPE WITH A SPECIFIC KEY TYPE AND VALUE TYPE

```typescript
const nameAgeMap: Record<string, number> = {
  'Alice': 21,
  'Bob': 25
};
console.log(nameAgeMap);
```

## omit -> removes key from object

```typescript
interface Person {
  name: string;
  age: number;
  location?: string;
}

const bob: Omit<Person, 'age' | 'location'> = {
  name: 'Bob'
  // `Omit` has removed age and location from the type and they can't be defined here
};
```

## Pick->removes all but the specified keys from an object type

```typescript
interface Person {
  name: string;
  age: number;
  location?: string;
}

const bob: Pick<Person, 'name'> = {
  name: 'Bob'
  // `Pick` has only kept name, so age and location were removed from the type and they can't be defined here
};
```

## exclude -> reove types from union

```typescript
type Primitive = string | number | boolean
const value: Exclude<Primitive, string> = true; // a string cannot be used here since Exclude removed it from the type.
```

## ReturnType extracts the return type of a function type.


```typescript
type PointGenerator = () => { x: number; y: number; };
const point: ReturnType<PointGenerator> = {
  x: 10,
  y: 20
};
```

## parameters-> extracts the parameter types of a function type as an array.

```typescript
type PointPrinter = (p: { x: number; y: number; }) => void;
const point: Parameters<PointPrinter>[0] = {
  x: 10,
  y: 20
};
```

## readonly


```typescript
interface Person {
  name: string;
  age: number;
}
const person: Readonly<Person> = {
  name: "Dylan",
  age: 35,
};
person.name = 'Israel'; // prog.ts(11,8): error TS2540: Cannot assign to 'name' because it is a read-only property.
```
